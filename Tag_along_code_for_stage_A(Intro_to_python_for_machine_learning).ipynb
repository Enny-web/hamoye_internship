Question 3 : What is the total protein supply in Madagascar in 2015?

```python
new_df = df[['Element','Area','Y2015']]
new_grouped_df = new_df.groupby(['Element','Area']).sum()
new_grouped_df
m_protein_supply_2015 = new_grouped_df.loc[('Protein supply quantity (g/capita/day)', 'Madagascar')]
print("The total Protein supply in Madagascar in 2015 is:",m_protein_supply_2015)
```
